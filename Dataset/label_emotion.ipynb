```python
import pandas as pd
import os
```

```python
data_info_df = pd.DataFrame(columns=['path','emotion','GT-STT'])
subject_path = '/home/neuroai/Desktop/LEE_JY/STT-Emotion_Model/Dataset/voice/'
subject_list = os.listdir(subject_path)
label_path = '/home/neuroai/Desktop/LEE_JY/STT-Emotion_Model/Dataset/label/'
label_list = os.listdir(label_path)
```

```python
subject_count = 0
for label in label_list:
    temp_df = pd.read_csv(label_path + label, encoding='cp949')
    for i in range(len(temp_df)):
        subject_id = temp_df.iloc[i]['wav_id'] + '.wav'
        gt_stt = temp_df.iloc[i]['발화문']
        temp_label = temp_df.iloc[i]['상황']
        if subject_id in subject_list:
            temp_sub_info = [{'path':subject_path + subject_id, 'emotion':temp_label, 'GT-STT':gt_stt}]
            df = pd.DataFrame.from_dict(temp_sub_info)
            data_info_df = pd.concat([data_info_df, df], ignore_index=True)
        else:
            print(f'error {subject_count}')
        subject_count+=1
```

```python
data_info_df['emotion'] = data_info_df['emotion'].replace({'anger': 'angry', 'sadness':'sad', 'happiness':'happy', 'fear':'fearful'})
```
